```python
import pandas as pd
import sqlite3
import os


db = sqlite3.connect("SoccerStats.db")

df = pd.read_csv("SoccerGameResults.csv")

df.to_sql("stats_import", db, if_exists = "replace")

def run_query(query):
    return pd.read_sql_query(query,db)

os.getcwd()
```

```python
query = """
SELECT *
FROM sqlite_master
ORDER BY name;
"""

run_query(query)
```

```python
query = """
SELECT *
FROM stats_import;
"""

run_query(query)
```
